#**Grayscale Image to RGB Image Conversion**

This document demostrates the conversion of an RGB Image to Grayscale image and then back to RGB.

###Steps:
1.   Original Image (RGB)
2.   Grayscale Image (RGB -> Grayscale)
3.   RGB Image (Grayscale -> RGB)

###Method
We used two different python libraries to demostrate the conversions and to confirm our results of the transformations.

Python libraries used:
1.   skimage
2.   OpenCV


###Conclusion
The image, once turned Grayscale can not be reverted into RGB.
This is because initally a RGB image has 3 channels (red, green, blue). Once it is changed to grayscale, the augmented image (grayscale image) has only 1 channel. That 1 channel can not be uniquely mapped to 3 channels (RGB).

**Using skimage**




```
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2gray 
from skimage.color import grey2rgb

original = data.astronaut()

# Convert Original image (RGB) to Grayscale
grayscale = rgb2gray(original)  # Y = 0.2125 R + 0.7154 G + 0.0721 B

# Convert Grayscale image to RGB
rgb = grey2rgb(grayscale) 

# Plot graph to hold 3 images
fig, axes = plt.subplots(1, 3, figsize=(12, 8))
ax = axes.ravel() # packs images next to eachother

# Original
ax[0].imshow(original)  #cmap ignored because 'original' has all 3 channels
ax[0].set_title("Original")

# Grayscale
ax[1].imshow(grayscale, cmap='gray') #maps augmented image to grayscale
ax[1].set_title("Grayscale")

# RGB
ax[2].imshow(rgb) #cmap ignored because 'rgb' has all 3 channels
ax[2].set_title("RGB")

fig.tight_layout()
plt.show()
```

**Using OpenCV**

```
import cv2
  
#image = cv2.imread('C:/Users/N/Desktop/Test.jpg') #loads local image
image = data.astronaut()  

# Convert Original image (RGB) to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert Grayscale image to RGB
image2 = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

# Plot graph to hold 3 images
fig, axes = plt.subplots(1, 3, figsize=(12, 8))
ax = axes.ravel() # packs images next to eachother

# Original
ax[0].imshow(original)  #cmap ignored because 'orignal' has all 3 channels
ax[0].set_title("Original")

# Grayscale
ax[1].imshow(gray, cmap='gray') #maps augmented image to grayscale
ax[1].set_title("Grayscale")

# Grayscale
ax[2].imshow(image2) #cmap ignored because 'image2' has all 3 channels
ax[2].set_title("RGB")
  
```
